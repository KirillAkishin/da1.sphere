```python
import numpy as np

import matplotlib.pyplot as plt

%matplotlib inline
```

```python
def complex_roots(n):
    z = -1
    k = np.array(range(n))
    pi = np.pi
    length = np.abs(z)
    
    sin_phi = np.imag(z) / length
    phi = np.arcsin(sin_phi)
    
    return length**(1/n) * np.exp((1j)*(phi+2*k*pi)/n)

roots = complex_roots(5)
roots
```

```python
roots = np.concatenate([roots, [roots[0]]])

plt.figure(figsize=(5, 5))
plt.xlim(-1.1, 1.1)
plt.ylim(-1.1, 1.1)
plt.plot(roots.real, roots.imag)
plt.show()
```

```python
def koch_curve():
    sin_alpha = np.sin(np.pi/3)
    a = 1/3
    height = sin_alpha * a

    return np.array([1, 2/3, 1/2 + height*(1j), 1/3, 0])
curve = koch_curve()
curve
```

```python
plt.figure(figsize=(5, 5))
plt.axis('equal')
plt.plot(curve.real, curve.imag)
plt.show()
```

```python
def change_edges_to_koch_curves(vertices):
    v = vertices
    n = len(v)
    v = np.repeat(v, 5, axis=0)
    
    pi = np.pi
    phi = - pi * 3/2

    alpha = 2/n * pi
    beta = (pi - alpha) / 2
    sin_beta = np.sin(beta)
    x = - sin_beta
    s = 2 * np.sqrt(1-x**2)

    curve = koch_curve()
#     c = (curve * np.exp((1j)*phi) - 1j/2) * s + x
    c = (curve - 1/2) * s - x*(1j)
    c = np.repeat([c], n, axis=0).ravel()

    line = c*v
    line = np.delete(line, range(0,n*5,5), axis=None) 
    return line
vertices = complex_roots(5)
line = change_edges_to_koch_curves(vertices)
line.shape
```

```python
line = np.concatenate([line, [line[0]]])

plt.figure(figsize=(5, 5))
plt.axis('equal')
plt.plot(line.real, line.imag)
plt.show()
```

```python
def koch_snowflake(order, scale=10):
    def _koch_snowflake_complex(order):
        if order == 0:
            vertices = complex_roots(3) * 1j
        else:
            vertices = _koch_snowflake_complex(order - 1)
            vertices = change_edges_to_koch_curves(vertices)
        return vertices

    points = scale * _koch_snowflake_complex(order)
    x, y = points.real, points.imag
    return x, y

x, y = koch_snowflake(order=2)

plt.figure(figsize=(8, 8))
plt.axis('equal')
plt.fill(x, y)
plt.show()
```
